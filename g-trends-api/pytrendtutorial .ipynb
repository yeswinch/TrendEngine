```python
import pandas as pd
from pytrends.request import TrendReq

pytrends = TrendReq()

pytrends.build_payload(kw_list=['Taylor Swift'])
```

```python
# Interest by Region
df = pytrends.interest_by_region()
df.head(10) #shows first 10 rows
```

```python
# Interest over Time
pytrends.interest_over_time() #shows first/last 5 rows
```

```python
#Plot Interest by Region in Bar Graph
df.reset_index().plot(x='geoName', y='Taylor Swift', figsize=(120, 10), kind ='bar')
```

```python
#Plot Interest by Region in Bar Graph
df.reset_index().plot(x='geoName', y='Taylor Swift', figsize=(120, 10), kind ='bar', resolution='ALGERIA')
```
